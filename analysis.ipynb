```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("sales_data.csv")

# Convert Sale_Date to datetime
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])

print(df.head())
```

```python
# Plot sales over time
daily_sales = df.groupby('Sale_Date')['Sales_Amount'].sum()

plt.figure()
plt.plot(daily_sales.index, daily_sales.values)
plt.xlabel("Date")
plt.ylabel("Total Sales Amount")
plt.title("Sales Over Time")
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("sales_over_time.png")
plt.show()
```

```python
# Monthly Sales
monthly_sales = df.resample('M', on='Sale_Date')['Sales_Amount'].sum()

plt.figure()
plt.plot(monthly_sales.index, monthly_sales.values)
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("monthly_sales.png")
plt.show()
```

```python
quarterly_sales = df.resample('Q', on='Sale_Date')['Sales_Amount'].sum()

plt.figure()
plt.plot(quarterly_sales.index, quarterly_sales.values)
plt.xlabel("Quarter")
plt.ylabel("Sales Amount")
plt.title("Quarterly Sales Trend")
plt.tight_layout()

plt.savefig("quarterly_sales.png")
plt.show()
```

```python
category_sales = df.groupby('Product_Category')['Sales_Amount'].sum()

plt.figure()
plt.bar(category_sales.index, category_sales.values)
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.title("Sales by Product Category")
plt.xticks(rotation=30)
plt.tight_layout()

plt.savefig("category_sales_bar.png")
plt.show()
```
